# Stop Word Removal


```python
#imports
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
#download stop words
import nltk
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stop_words
```

```python
#Example Sentence
example = "I alone cannot change the world, but I can cast a stone across the water to create many ripples."
```

```python
#Tokenize Sentence
nltk.download('punkt')
tokens = word_tokenize(example)
tokens
```

```python
# filter sentence
filtered_text = [word for word in tokens if not word in stopwords.words()]
filtered_text
```

# Stemming


```python
# import modules
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# create object 
ps = PorterStemmer()

# take different variations of the same word
words = ["wait", "waiting", "waited", "waits"]

# apply ps.stem 
for w in words:
 print(w, " : ", ps.stem(w))
```

```python
# Apply stem on filtered_text(example data)
for word in filtered_text:
  rootWord=ps.stem(word)
  print(rootWord)
```

# Lemmatization


```python
# import modules
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
```

```python
# lemmatization examples
rock = lemmatizer.lemmatize("rocks");
apple = lemmatizer.lemmatize("apples")
print("rocks :", rock)
print("apples :", apple)

# a denotes adjective in "pos"
print("better :", lemmatizer.lemmatize("better", pos ="a"))
```

```python
# Apply Lemmatization on filtered text(example statement)
for word in filtered_text:
  rootWord=lemmatizer.lemmatize(word, pos ="a")
  print(rootWord)
```
